### **STANDARDIZE ALL IMAGES NOW , first reading with listdir , then asking to user for choices of cropping**

```python
import cv2
import os
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
from matplotlib import pyplot as plt
choice=input("Enter your preference(Original/Resized/Cropped/Filled): ")
stack_of_images=[]

# ORIGINAL
if(choice=="Original"):
    for filename in os.listdir("/content/gdrive/MyDrive/Aliens"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Aliens",filename))
        if img is not None:
            plt.imshow(img)
            plt.axis('off')
            plt.show()     
            stack_of_images.append(img)
  
    for filename in os.listdir("/content/gdrive/MyDrive/Zombie"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Zombie",filename))
        if img is not None:
            plt.imshow(img)
            plt.axis('off')
            plt.show()
            stack_of_images.append(img)
   
#RESIZED
elif(choice=="Resized"):
    for filename in os.listdir("/content/gdrive/MyDrive/Aliens"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Aliens",filename))
        if img is not None:
            img_resized = cv2.resize(img, (200,200))
            plt.imshow(img_resized)
            plt.axis('off')
            plt.show()
            stack_of_images.append(img_resized)

    for filename in os.listdir("/content/gdrive/MyDrive/Zombie"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Zombie",filename))
        if img is not None:
            img_resized = cv2.resize(img, (200,200))
            plt.imshow(img_resized)
            plt.axis('off')
            plt.show()
            stack_of_images.append(img_resized)

# CROPPED (MANUAL INPUT FOR WIDTH AND HEIGHT)
elif(choice=="Cropped"):
    w = int(input("Enter Width : "))
    h = int(input("Enter Height : "))
    for filename in os.listdir("/content/gdrive/MyDrive/Aliens"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Aliens",filename))
        if img is not None:
            w=img.shape[1]
            h=img.shape[0]
            im1 = img[((h-100)//2):((h-100)//2)+((h+100)//2), ((w-100)//2):((w-100)//2)+((w+100)//2)]
            plt.imshow(im1)
            plt.axis('off')
            plt.show()
            stack_of_images.append(im1)
    for filename in os.listdir("/content/gdrive/MyDrive/Zombie"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Zombie",filename))
        if img is not None:
            w=img.shape[1]
            h=img.shape[0]
            im1=img[((h-100)//2):((h-100)//2)+((h+100)//2), ((w-100)//2):((w-100)//2)+((w+100)//2)]
            plt.imshow(im1)
            plt.axis('off')
            plt.show()
            stack_of_images.append(im1)

# FILLED
elif(choice=="Filled"):
    for filename in os.listdir("/content/gdrive/MyDrive/Aliens"):
        im = cv2.imread(os.path.join("/content/gdrive/MyDrive/Aliens",filename))
        if im is not None:
            result=cv2.copyMakeBorder(im, 0, 0, 200, 200, cv2.BORDER_CONSTANT, None, value = 0)
            plt.imshow(result)
            plt.axis('off')
            plt.show()
            stack_of_images.append(result)
    for filename in os.listdir("/content/gdrive/MyDrive/Zombie"):
        im = cv2.imread(os.path.join("/content/gdrive/MyDrive/Zombie",filename))
        if im is not None:
            result=cv2.copyMakeBorder(im, 0, 0, 200,200, cv2.BORDER_CONSTANT, None, value = 0)
            plt.imshow(result)
            plt.axis('off')
            plt.show()
            stack_of_images.append(result)
```

### **GREYSCALE IMAGES Dataset**

```python
import cv2
import os
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
from matplotlib import pyplot as plt
greyscale=[]
for filename in os.listdir("/content/gdrive/MyDrive/Aliens"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Aliens",filename))
        if img is not None:
            if(choice=="Original"):
                gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Resized"):
                img_resized = cv2.resize(img, (200,200))
                gray_image = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Cropped"):
                w=img.shape[1]
                h=img.shape[0]
                im1 = img[((h-100)//2):((h-100)//2)+((h+100)//2), ((w-100)//2):((w-100)//2)+((w+100)//2)]
                gray_image = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Filled"):
                result=cv2.copyMakeBorder(img, 0, 0, 200, 200, cv2.BORDER_CONSTANT, None, value = [255,255,255])
                gray_image = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
for filename in os.listdir("/content/gdrive/MyDrive/Zombie"):
        img = cv2.imread(os.path.join("/content/gdrive/MyDrive/Zombie",filename))
        if img is not None:
            if(choice=="Original"):
                gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Resized"):
                img_resized = cv2.resize(img, (200,200))
                gray_image = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Cropped"):
                w=img.shape[1]
                h=img.shape[0]
                im1 = img[((h-100)//2):((h-100)//2)+((h+100)//2), ((w-100)//2):((w-100)//2)+((w+100)//2)]
                gray_image = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
            elif(choice=="Filled"):
                result=cv2.copyMakeBorder(img, 0, 0, 200, 200, cv2.BORDER_CONSTANT, None, value = [255,255,255])
                gray_image = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
                plt.imshow(gray_image,cmap='gray')
                plt.axis('off')
                plt.show()
                greyscale.append(gray_image)
```

### **BINARY IMAGES Dataset**
