# **TEXT FILE**

```python
import glob
alltxtfiles=[]
alltxtfiles = glob.glob("/content/text/*")
print(alltxtfiles[1])
```

```python
import re
combinationofdatas={}
txtdatas1=[]
sum=0

for i in range(len(alltxtfiles)):
  
  with open(alltxtfiles[i],'r') as file:    
    for line in file:  
      for word in line.split():
        word1 = re.sub(r'[^a-zA-Z0-9]',"",word)
        word2 = word1.lower()
        txtdatas1.append(word2)

  combinationofdatas[alltxtfiles[i]] = txtdatas1[sum:sum+len(txtdatas1)]
  sum=sum+len(combinationofdatas[alltxtfiles[i]])

print((combinationofdatas[alltxtfiles[0]]))
```
